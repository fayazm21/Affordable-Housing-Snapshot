```python
import pandas as pd
import matplotlib.pyplot
import seaborn as sns
import warnings 

warnings.filterwarnings("ignore")
```

```python
#cleaning and combining the 311 datasets:
#make sure you have the filenames renamed correctly to these for this cell to execute
threeOneOneRequests = ['datasets/311_service_requests_2017.csv','datasets/311_service_requests_2018.csv','datasets/311_service_requests_2019.csv','datasets/311_service_requests_2020.csv','datasets/311_service_requests_2021.csv','datasets/311_service_requests_2022.csv']

df_311 = pd.read_csv('datasets/311_service_requests_2016.csv')
for i in threeOneOneRequests:
	temp_df = pd.read_csv(i)
	df_311 = pd.concat([df_311, temp_df])

print(df_311.shape) #to check the file size before dropping 
#dropping columns
df_311.drop(['queue','department', 'submittedphoto','closedphoto','fire_district','neighborhood_services_district','precinct','ward'], axis =1)

#retaining reasons in the reason column that closely match with rental violations
df_311 = df_311[df_311['reason'].isin(['Building','Housing','Code Enforcement','Recycling','Noise Disturbance','Generic Noise Disturbance','Neighborhood Services Issues','Parking Complaints','Environmental Services'])]

print(df_311.shape) #to check the file size after dropping
```

```python
#adding a new column, which will be empty at first

df_311['violation_type'] = ''

df_311
```

```python
import numpy as np
#matching the reasons with the violation types from rent smart.
#we matched the closest reason to the 6 types of violation types from rent smart to have some consistency
#
conditions = [(df_311['reason'] == 'Code Enforcement'), # will match to Enforcement Violation
              
              (df_311['reason'] == 'Noise Disturbance'),
              (df_311['reason'] == 'Generic Noise Disturbance'), # Will match to Housing violations
              
              (df_311['reason'] == 'Recycling'),
              (df_311['reason'] == 'Environmental Services'),# will match to sanitation requests
              
              (df_311['reason'] == 'Neighborhood Services Issues'),
              (df_311['reason'] == 'Parking Complaints'), # will match to civic maintanance requests
              
              (df_311['reason'] == 'Building'), #will match to building violations
              
              (df_311['reason'] == 'Housing') #will match to housing violations 
             ]

#violation types to be matched with
violation_types = ['Enforcement Violations',
                   
                   'Housing Violations','Housing Violations',
                   
                   'Sanitation Requests','Sanitation Requests',
                   
                   'Civic Maintenance Requests', 'Civic Maintenance Requests',
                   
                   'Building Violations', 
                   
                   'Housing Complaints'
                  ]

df_311['violation_type'] = np.select(conditions, violation_types)

df_311
```

```python
#changing the 'location_street_name' to string type
df_311['location_street_name'] = df_311['location_street_name'].astype(str)
```

```python
#converting to uppercase letters
df_311['location_street_name'] = df_311['location_street_name'].apply(lambda x: x.upper())
```

```python
print(parcels.columns)
print(df_311.columns)
```

```python
#dropping unnecessary columns
parcels.drop(['mapc_id', 'muni_id' ,'muni', 'parloc_id', 'poly_typ',
              'map_num', 'mappar_id', 'loc_id_cnt', 'land_value', 'bldg_value','othr_value','sqm_imperv',
              'sqm_bldg', 'sqm_pave','realesttyp', 'temp', 'shape', 'units_src', 'num_rooms', 'fy', 'lot_areaft', 'far', 
              'pct_imperv', 'pct_bldg', 'pct_pave', 'landv_pac', 'bldgv_psf','totv_pac', 'bldlnd_rat',
              'addr_str', 'addr_num', 'owner_stat', 'addr_zip', 'num_units', 'units_est', 'luc_1', 'luc_2'
              , 'luc_adj_1', 'luc_adj_2', 'objectid'], axis = 1, inplace = True)
```

```python
# converting ls_date to str type
parcels['ls_date'] = parcels['ls_date'].astype(str)
```

```python
# restructured ls_date, previously it was in integer form 
parcels['ls_date'] = parcels['ls_date'].apply(lambda x: (x[:4] + '/' + x[4:6] + '/' + x[6:9]))
```

```python
# removing the rows having NaN values, we have enough data so its better dropping the NaN rows as we cant replace the zip codes
import numpy as np
parcels = parcels[~parcels['owner_zip'].isin([np.NaN])]
parcels = parcels[~parcels['owner_city'].isin([np.NaN])]
parcels = parcels[~parcels['owner_addr'].isin([np.NaN])]
parcels = parcels[~parcels['owner_name'].isin([np.NaN])]
parcels = parcels[~parcels['site_addr'].isin([np.NaN])]
parcels = parcels[~parcels['yr_built'].isin([np.NaN])]
```

```python
# renaming the 'site_addr column' to 'address' so that we can merge addresses under single column
parcels = parcels.rename(columns={'site_addr': 'address'})
```

```python
# renaming the location_street_name column to address so that we can merge addresses under single column
df_311 = df_311.rename(columns={'location_street_name': 'address'})
```

```python
# merging df_311, parcels on address (innerjoint) 
merged_311 = pd.merge(df_311, parcels, on = 'address', how = 'inner')
```
