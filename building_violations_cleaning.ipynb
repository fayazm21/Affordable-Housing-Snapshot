```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
import seaborn as sns
import warnings 

warnings.filterwarnings("ignore")
```

```python
#import and view build violations data
build_violations = pd.read_csv("datasets/building_violations.csv")
build_violations.head(4)
```

```python
#adding the violation type column and Building violation values. 
#Since the whole df describes building violations, they can all fall under building violation when matching with rentsmart

build_violations['violation_type'] = 'Building Violations'

build_violations 
```

```python
build_violations.columns
```

```python
#make an address column, make it upper case
build_violations['address'] = build_violations['violation_stno']+' '+build_violations['violation_street']+' '+build_violations['violation_suffix']
build_violations['address'] = build_violations['address'].astype(str)
build_violations['address'] = build_violations['address'].apply(lambda x: x.upper())
```

```python
#import parcel data
parcels = pd.read_csv('datasets/parcel.csv')
print(parcels.columns)
parcels.head()
```

```python
#clean parcel data before merge
parcels.drop(['mapc_id', 'muni_id' ,'muni', 'parloc_id', 'poly_typ',
              'map_num', 'mappar_id', 'loc_id_cnt', 'land_value', 'bldg_value','othr_value','sqm_imperv',
              'sqm_bldg', 'sqm_pave','realesttyp', 'temp', 'shape', 'units_src', 'num_rooms', 'fy', 'lot_areaft', 'far', 
              'pct_imperv', 'pct_bldg', 'pct_pave', 'landv_pac', 'bldgv_psf','totv_pac', 'bldlnd_rat',
              'addr_str', 'addr_num', 'owner_stat', 'addr_zip', 'num_units', 'units_est', 'luc_1', 'luc_2'
              , 'luc_adj_1', 'luc_adj_2', 'objectid'], axis = 1, inplace = True)
parcels['ls_date'] = parcels['ls_date'].astype(str)
parcels['ls_date'] = parcels['ls_date'].apply(lambda x: (x[:4] + '/' + x[4:6] + '/' + x[6:9]))
parcels = parcels[~parcels['owner_zip'].isin([np.NaN])]
parcels = parcels[~parcels['site_addr'].isin([np.NaN])]
parcels = parcels[~parcels['owner_name'].isin([np.NaN])]
parcels = parcels[~parcels['owner_addr'].isin([np.NaN])]
parcels = parcels[~parcels['owner_city'].isin([np.NaN])]
parcels = parcels[~parcels['yr_built'].isin([np.NaN])]
parcels = parcels.rename(columns={'site_addr': 'address'})
parcels.shape
```

```python
merged_building_vio = pd.merge(build_violations, parcels, how='inner', on = 'address')
merged_building_vio.shape
```
